```python
from ipyleaflet import Map, basemaps, basemap_to_tiles, Polygon
```

```python
# Pleiades NEO coverage

#General Location
center = [43.610052, 5.113754]
zoom = 11

# Pleiades NEO MS imagery from the .xml file
polygon = Polygon(
    locations=[(43.673678628297594,4.9767047783854705), 
               (43.66274143391238,5.246936341920053), 
               (43.561931402913295,5.245880797182931),
               (43.57121627179229,4.97705377747287)],
    
    color="red",
    opacity=0.1,
)

# Pleiades NEO PAN imagery from the .xml file
polygon1 = Polygon(
    locations=[(43.67368580266746,4.976693024793952), 
               (43.662747363387865,5.246949511818026), 
               (43.5619246358925,5.245890391135575),
               (43.57121004594763,4.977042016909566)],
    
    color="blue",
    opacity=0.1,
)


m = Map(basemap=basemaps.OpenStreetMap.France, center=center, zoom=zoom)
m.add(polygon);
m.add(polygon1);

m
```
