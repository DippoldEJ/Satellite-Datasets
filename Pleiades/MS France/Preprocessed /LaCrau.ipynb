```python
import ipyleaflet
from ipyleaflet import Map, basemaps, basemap_to_tiles, Marker, AwesomeIcon, Polyline
```

```python
# Location and zoom level  
center = [43.152099, 6.078253] 
zoom = 13

m = Map(basemap=basemaps.OpenStreetMap.Mapnik, center=center, zoom=zoom)

icon1 = AwesomeIcon(
    name='space-shuttle',
    marker_color='red',
    icon_color='black',
    spin=False
)

marker1 = Marker(icon=icon1, location=(43.150917, 6.074262))
m.add(marker1)

# lets check the current map
m
```
