```python
from ipyleaflet import Map, basemaps, basemap_to_tiles, Polygon
```

```python
# La Grau Pleiades coverage

#General Location
center = [43.610052, 5.113754]
zoom = 11

# Precise Location & Coverage of the Pleiades MS imagery from the .xml file
polygon = Polygon(
    locations=[(43.65608885545679,5.05103021813204), 
               (43.66173915021084, 5.188630638685853), 
               (43.55583515192971, 5.188686886952479),
               (43.54982299396757, 5.051762810207236)],
    
    color="blue",
    fill_color="blue"
)


m = Map(basemap=basemaps.OpenStreetMap.France, center=center, zoom=zoom)

m.add(polygon);

m
```
